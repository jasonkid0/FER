# real all images and convert them into array

```python
training_Data = [] ##data array

def create_training_Data():
    for category in Classes:
        path = os.path.join(Datadirectory, category)
        class_num = Classes.index(category)
        for img in os.listdir(path) :
            try:
                img_array = cv2.imread(os.path.join(path,img))
                new_array = cv2.resize(img_array, (img_size,img_size))
                training_Data.append([new_array, class_num])
            except Exception as e:
                pass
```

```python
create_training_Data()
```

```python
print(len(training_Data))
```

# we are just using training for webcam demo

```python
import random

random.shuffle(training_Data)
```

```python
X = [] ## data /feature
y = [] ## label

for features, label in training_Data:
    X.append(features)
    y.append(label)
    
X = np.array(X).reshape(-1, img_size, img_size, 3) ## converting it to 4 dimension
```

```python
X.shape
```

```python
# normalize the data
X = X/255.0; ## normalizing the dataset
```

```python
Y =np.array(y)
```

# deep learning model for training - Transfer Learning

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.MobileNetV2()  ## Pre-trained Model
```

```python
model.summary()
```

# Transfer Learning - Tuning ,weights will start from the last checkpoint

```python
base_input = model.layers[0].input ## input
```

```python
base_output = model.layers[-2].output
```

```python
base_output
```

```python
final_output =layers.Dense(128)(base_output) ## adding new layer , after the output of global pooling layer
final_output = layers.Activation('relu')(final_output) ## activation function
final_output = layers.Dense(64)(final_output) 
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(7,activation='softmax')(final_output) ## my classes are 07,classification layer
```

```python
final_output ## output
```

```python
new_model = keras.Model(inputs = base_input, outputs = final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss="sparse_categorical_crossentropy",optimizer = "adam", metrics = ["accuracy"])
```

```python
new_model.fit(X,Y, epochs = 25)
```

```python
new_model.save('Final_model_95p07.h5')
```

```python
    import tensorflow as tf ##pip install tensorflow
    import cv2 ##pip install opencv-python
    import os
    import matplotlib.pyplot as plt ##pip install matplotlib
    import numpy as np ##pip install numpy


    # Path to the Haar Cascade classifier XML file
    path = "haarcascade_frontalface_default.xml"

    # Font settings for displaying emotion labels
    font_scale = 1.5
    font = cv2.FONT_HERSHEY_PLAIN

    # Initialize a white background image with text
    rectangle_bgr = (255, 255, 255)
    img = np.zeros((500, 500, 3), dtype=np.uint8)
    text = "Some text in a box!"

    # Get the size of the text box
    (text_width, text_height), _ = cv2.getTextSize(text, font, fontScale=font_scale, thickness=1)

    # Set the text start position and the box coordinates
    text_offset_x = 10
    text_offset_y = img.shape[0] - 25
    box_coords = ((text_offset_x, text_offset_y), (text_offset_x + text_width + 2, text_offset_y - text_height - 2))

    # Draw the text box and text on the image
    cv2.rectangle(img, box_coords[0], box_coords[1], rectangle_bgr, cv2.FILLED)
    cv2.putText(img, text, (text_offset_x, text_offset_y), font, fontScale=font_scale, color=(0, 0, 0), thickness=1)

    # Open the webcam
    cap = cv2.VideoCapture(0)

    if not cap.isOpened():
        raise IOError("Cannot open webcam")

    # Load the face cascade classifier
    faceCascade = cv2.CascadeClassifier(path)

    # Define emotion labels
    emotion_labels = ["Angry", "Disgust", "Fear", "Happy", "Neutral", "Sad", "Surprise"]

    while True:
        ret, frame = cap.read()

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        faces = faceCascade.detectMultiScale(gray, 1.1, 4)

        for x, y, w, h in faces:
            roi_gray = gray[y:y + h, x:x + w]
            roi_color = frame[y:y + h, x:x + w]

            final_image = cv2.resize(roi_color, (224, 224))
            final_image = np.expand_dims(final_image, axis=0)
            final_image = final_image / 255.0

            # Predict emotion using the emotion model (replace 'new_model' with your actual model)
            Predictions = new_model.predict(final_image)
            predicted_emotion = emotion_labels[np.argmax(Predictions)]

            # Draw a rectangle around the detected face and display the predicted emotion
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)
            cv2.putText(frame, predicted_emotion, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)

        # Display the frame with emotions
        cv2.imshow('Face Emotion Recognition', frame)

        # Break the loop when 'q' is pressed
        if cv2.waitKey(2) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```
